```python
#Libraries
import graphviz
import xgboost as xgb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
```

```python
X, y = make_blobs(n_samples=2000, n_features=3, random_state=42)
X.shape
```

```python
Xdf = pd.DataFrame(X)
yseries = pd.Series(y)
```

```python
Xdf.info()
```

```python
Xdf.describe().T
#Also there is not any missing value.
```

```python
yseries.value_counts()
#Y value consists of variables 0,1 and 2.
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x=0, y=1, hue=2, data=Xdf);
#As we can see in the graph, our data is in a classification among itself.
```

```python
sns.pairplot(Xdf);
#We observe the relationships.
```

```python
#Now I am examining whether we have outlier data.
z = np.abs(stats.zscore(Xdf))
outliers = list(set(np.where(z > 3)[0]))
categories = list(([0, 1, 2]))
len(outliers)
#As we saw, we havent got any outlier data.
```

```python
#Split our data as train and test.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train,y_train)
print("Accuracy of train:",clf.score(X_train,y_train))
print("Accuracy of test:",clf.score(X_test,y_test))
#We got a very good result due to the distribution of our data.
```

```python
dot_data = export_graphviz(clf, out_file=None,
                     feature_names=Xdf.columns,
                     filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph
```

```python
# Classification Report
pred = clf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
# Metrics
print("Precision = {}".format(precision_score(y_test, pred, average='macro')))
print("Recall = {}".format(recall_score(y_test, pred, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, pred)))
print("F1 Score = {}".format(f1_score(y_test, pred,average='macro')))
```

```python
dmatrix_train = xgb.DMatrix(data=X_train, label=y_train)
dmatrix_test = xgb.DMatrix(data=X_test, label=y_test)
param = {'max_depth':3, 
         'eta':1, 
         'objective':'multi:softprob', 
         'num_class':3}

num_round = 5
model = xgb.train(param, dmatrix_train, num_round)
```

```python
preds = model.predict(dmatrix_test)
preds[:10]
```

```python
best_preds = np.asarray([np.argmax(line) for line in preds])
print("Precision = {}".format(precision_score(y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, best_preds)))
```
